```python
import pandas as pd
import numpy as np
import timeit
```

```python
#===========================Phrases===============================

# NB: The phrases dataset has the data encoded into utf-8, but sentences dataset is encoded into latin-1
# To balance these out, I must convert phrase dataset to latin-1 encoding, in order to have the strings 
# to match each other
```

```python
sentiment_labels = pd.read_table("sentiment_labels.txt", sep = "|", encoding = "latin-1")
sentiment_labels = sentiment_labels.rename({"phrase ids": "id", "sentiment values": "values"}, axis = 1)
sentiment_labels
```

```python
dictionary = pd.read_table("dictionary.txt", sep = "|", header = None, names = ["phrase", "id"], encoding = "latin-1")
dictionary
```

```python
df_phrase = pd.merge(sentiment_labels, dictionary, on = "id")
df_phrase
```

```python
#==========================Sentences===========================
```

```python
datasetSentences = pd.read_table("datasetSentences.txt", sep = "\t", encoding = "utf-8")

for index, sent in datasetSentences.iterrows():
    newSent = datasetSentences["sentence"][index].replace("-LRB-", "(")
    newSent = newSent.replace("-RRB-", ")")
    datasetSentences["sentence"][index] = newSent

datasetSentences
```

```python
datasetSplit = pd.read_table("datasetSplit.txt", sep = ",")
datasetSplit
```

```python
# Sentences with the index and its corresponding label (1 = train, 2 = dev, 3 = test)
df_sentences = pd.merge(datasetSentences, datasetSplit, on = "sentence_index")
df_sentences
```

```python
#==============Splitting dataset into train, dev, test=============
```

```python
df_sent_train = df_sentences[df_sentences["splitset_label"] == 1]
df_sent_dev = df_sentences[df_sentences["splitset_label"] == 3]
df_sent_test = df_sentences[df_sentences["splitset_label"] == 2]
```

```python
df_sent_train
```

```python
df_sent_dev
```

```python
df_sent_test
```

```python
# Check if the sentences dataset is the subset of phrases dataset
phrases = df_phrase["phrase"].values
sentences = df_sentences["sentence"].values

set(sentences).issubset(set(phrases))
```

```python
# The above value returned True, which means that 
# phrases dataset is including all sentences in train, dev, and test sets
# We now try to partition these sentences with the corresponding classes
```

```python
# Make a copy of original df_phrase dataset, so that we can start working with the copy of the original dataset
df_phrase_sent = df_phrase.rename({"phrase": "sentence"}, axis = 1)
df_phrase_sent
```

```python
df_train = pd.merge(df_sent_train, df_phrase_sent, on = "sentence")
df_train
```

```python
df_test = pd.merge(df_sent_test, df_phrase_sent, on = "sentence")
df_test
```

```python
df_dev = pd.merge(df_sent_dev, df_phrase_sent, on = "sentence")
df_dev
```

```python
# Now, we have three datasets available:
#     - df_train: training set
#     - df_dev: development set
#     - df_test: testing set
    
#     All the datasets have its sentence, and its corresponding values
```

```python
# Create separate CSV file for each dataset for future reuse
df_train.to_csv("SSTB_train.csv")
df_dev.to_csv("SSTB_dev.csv")
df_test.to_csv("SSTB_test.csv")
```

```python
# Getting phrases
# Make sure to eliminate the dev set and test set from the phrase list
# as we have previously discovered that both these sets are the subsets of
# the entire phrases
```

```python
# Removing dev set
for index, dev in df_dev.iterrows():
    df_phrase_sent = df_phrase_sent.drop(df_phrase_sent.index[df_phrase_sent.id == dev.id])
df_phrase_sent

```

```python
# Removing test set
for index, tt in df_test.iterrows():
    df_phrase_sent = df_phrase_sent.drop(df_phrase_sent.index[df_phrase_sent.id == tt.id])
df_phrase_sent
```

```python
# Save df_phrase_sent as a separate csv file for future reuse
df_phrase_sent.to_csv("SSTB_phrases_without_sentences.csv")
```
